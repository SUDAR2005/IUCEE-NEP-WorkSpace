Artificially generating dataset for Linear Model

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
#generating dataset
np.random.seed(0)
x = np.random.rand(100, 1)
noise = np.random.rand(100, 1)
y = 2 + 3*x + noise
dataset = pd.DataFrame(data={'x': x.flatten(), 'y': y.flatten()})
dataset.head()
```

Draw a plot to represent the dataset

```python
#creating plot
plt.scatter(x,y, s = 20) 
plt.xlabel('x (random value btw (0,1))') 
plt.ylabel('y (with noise)') 
plt.show()
```

```python
#splitting dataset using buitin class 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=2)
```

```python
#Create lear regression object
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
#train the Model
regr.fit(x_train, y_train)
#make predictions
y_predict = regr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

#analysing result
mse = mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
print("Coefficients: ", regr.coef_)
#the mean squared error
print(f"Mean squared error: {mean_squared_error(y_test,y_predict)}" )
#the coefficient of determination
print(f"Coefficient of determination: {r2_score(y_test, y_predict)}" )


plt.scatter(x,y, s = 20) 
plt.xlabel('x (random value btw (0,1))') 
plt.ylabel('y (with noise)') 
plt.plot(x_test,y_predict,color='green',markersize=0,linewidth=2)
plt.show()
```
